###What is concrete?

The basic composition of concrete is cement, gravel, and sand. It’s a very important ‘liquid rock’ that is very easy to use and inexpensive (“Explaining concrete”).
Cement and concrete are not the same thing. Cement is one of the ingredients in concrete while concrete itself is a mixture of aggregates and paste (“FAQs”). Aggregates are sand and gravel or crushed stone; the paste is water and cement (“FAQs”). The three basic ingredients in modern-day conventional concrete are Portland cement, fine and coarse aggregates, and water (Yeh 1998).

Cement is 10-15% of the concrete mix (by volume) and through hydration, the cement and water harden and bind the aggregates into a rocklike mass (“FAQs”). To put it real simple, the water reacts with the limestone to form calcium hydroxide which sticks all the aggregates together like glue. This process continues for years - concrete gets stronger as it gets older (“FAQs”).

Cement is the glue of concrete. It’s a very important substance that we use an incredible amount of (“Explaining concrete”) and is basically powder from which mortar and concrete are made (Crawford 2018).

Limestone has high calcium content and is mostly made up of CaCO3 (calcium carbonate) (Crawford 2018). When heated to about 1000 degrees celcius it produces quicklime (powder) which removes the carbon dioxide and only leaves calcium oxide: CaCO3 + heat = CaO + CO2 (Crawford 2018). You mix the quicklime with water ((H2O) which gives us calcium hydroxide (Ca(OH)2) (“Explaining concrete”). Calcium hydroxide - otherwise known as slake or hydrated lime - is then poured into a mold and absorbs CO2 from the air to return to calcium carbonate and completes the lime cycle (Crawford 2018). However, this is the most basic, primitive form of cement (“Explaining concrete”). You can’t use this for big projects as CO2 wouldn’t be able to penetrate all of the mixture and of course cannot be used underwater due to the lack of CO2 (“Explaining concrete”). It is too brittle and weak to be useful so we strengthen it by adding sand, gravel, and pebbles called aggregates (Crawford 2018). You can’t only add sand to it either as we would just end up with mortar (the glue that holds bricks/stones together) (Crawford 2018).

Portland cement is the most widely used cement (Crawford 2018). What separates it from the rest is that it’s made from clinkers - basically very tough rocks that are produced when you overcook lime and clay mixtures to real high temperatures - and we get tricalcium silicate in the process (“Explaining concrete”).


High-performance concrete, which is the type of concrete all of the samples in the data are, needs supplementary cementitious materials: fly ash, blast furnace slag, and chemical admixtures (superplasticizers).

Aggregates are another part of the formula. These are mostly sand and gravel which are blasted out of a quarry and ground down to particular sizes and shapes (which affects strength) (“Explaining concrete”). These are broadly categorized as ‘fine’ and ‘coarse’ aggregates, which can be understood as particles that are smaller than 5 mm and particles larger than 5 mm. In general, fine aggregates have more surface area, so more fine aggregate, in theory, will be stronger.

Superplasticizers are chemicals added to concrete that make it easier to work with and spread around without changing the water content (“Explaining concrete”). If you add water, it will affect the strength so superplasticizers are used instead (“Explaining concrete”).

Fly ash is finely divided residue that results from the combustion of pulverized coal and is transported from the combustion chambers through exhaust gas - basically a coal byproduct (“Fly Ash”). It’s used in Portland cement concrete (PCC), usually as a pozzolan and increases the life of concrete by improving its durability (“Fly Ash”).




## Data collection

Data is sourced from UCI Machine Learning Repository contributed by Prof Yeh of Chung-Hua University.<br>
There were about 1000 samples but some were deleted due to larger size aggregates (larger than 20 mm), special curing conditions, etc. 700 samples with ordinary Portland cement and cured under normal conditions were evaluated. The specimen types were converted into 15 cm cylinders through accepted guidelines and the mixing was controlled in a laboratory. Each quoted strength value is the average of the strengths from five cylinders (Yeh 1998).


###Testing

Concrete samples are made into cylinders of specific size and its compression strength is tested at 7, 14, and 28 days, though it continues to strengthen after that  (“Explaining concrete”). Early strength is measured at 7 days and concrete is nominally at ‘full strength’ after 28 days (Crawford 2019). 28 days was selected as the age all concrete should be tested as a substantial percent of the hydration process has taken place by this time (“FAQs”).

Conventional concrete is able to take 7,000 psi or less and high-strength concrete is able to take 7,000 to 14,400 psi (“FAQs”).
- Psi = pounds per square inch, compressive strength of concrete

### Feature Engineering



We do feature engineering to improve the model by increasing the likelihood that one of the new engineered variable would have more predictive power in predicting concrete strength.
- Aggregate = Total amount of fine and coarse aggregate
- Cement-water ratio = Cement/Water
- Coarse and Fine Aggregate Ratio = Coarse aggregate + Fine aggregate
- Aggregate-water ratio = Aggregate/Water

However, we have to be careful of overfitting in our model when using engineered features as they are calculated with the original variables meaning features are going to be highly correlated to each other.
- We can consider LASSO regression to reduce multicollinearity problem.

```python
data["aggregate"] = data["fineaggregate"] + data["coarseaggregate"]
data["cement/water"] = data["cement"]/data["water"]
data["coarse/fine"] = data["coarseaggregate"]/data["fineaggregate"]
data["aggregate/water"] = data["aggregate"]/data["water"]
```

###Organise data by approximate age

The effect of age of concrete on the concrete strength is dependent on the composition largely. We know from research that the strength of concrete tends to stagnate after 28 days. Hence, the role age plays on concrete strength is still largely dependent on the composition in the first place.

**To investigate how the difference in composition in concrete affects concrete strength, we can consider removing the effect of age.** We will filter data such that each dataframe contains data from the same period.

From research, concrete is tested in fixed intervals usually. Our data sees values like 91 and 360 days which are approximated to be 90 and 365 days respectively for simplicity.

```python
data["approx_age"] = data.loc[:,"age"]
data.loc[data["approx_age"]==360, "approx_age"] = 365
data.loc[data["approx_age"]==91, "approx_age"] = 90
data["approx_age"].value_counts().sort_index()
```

```python
from sklearn.model_selection import KFold

rmse_dic_lr = {}
r2_dic_lr = {}

for n_splits in range(2,11):
  kf = KFold(n_splits = n_splits, shuffle = True)
  rmse_lr = 0
  r2_lr = 0
  for i, (train_index, test_index) in enumerate(kf.split(X)):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    rmse_temp = np.sqrt(mean_squared_error(y_test, y_pred))
    r2_temp = r2_score(y_test, y_pred)
    rmse_lr += rmse_temp
    r2_lr += r2_temp
  r2_lr /= n_splits
  rmse_lr /= n_splits
  rmse_dic_lr[n_splits] = rmse_lr
  r2_dic_lr[n_splits] = r2_lr
```

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LassoCV

n_splits = list(rmse_dic_lr.keys())[0]
kf = KFold(n_splits=n_splits, shuffle = True)
rmse_dic_lasso = {}
r2_dic_lasso = {}

for cv in range(2,11):
  r2_lasso = 0
  rmse_lasso = 0
  for i, (train_index, test_index) in enumerate(kf.split(X)):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    scaler = StandardScaler()
    scaler.fit(X_train)
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)

    lasso = LassoCV(alphas = None, cv = cv, max_iter = 100000)
    lasso.fit(X_train, y_train)
    y_pred = lasso.predict(X_test)
    rmse_temp = np.sqrt(mean_squared_error(y_test, y_pred))
    r2_temp = r2_score(y_test, y_pred)
    r2_lasso += r2_temp
    rmse_lasso += rmse_temp
  r2_lasso = r2_lasso / n_splits
  rmse_lasso = rmse_lasso / n_splits

  rmse_dic_lasso[cv] = rmse_lasso
  r2_dic_lasso[cv] = r2_lasso
```

##Conclusion and Future Work

###Limitations
*   Insufficient data
*   The data/research we based this off is outdated (from 1998)
*   There are too many factors attributed to concrete strength which makes it difficult to generalize and make an accurate model off. Concrete mixes vary and also change on the situation. Ex) lighter concrete for certain structures or heavier, stronger concerete
*   Each component of concrete is described as a single term but it represents a variety of forms. Cement can be powdered to various degrees and have different chemical compositions. Component types also influence strength but the mixing proportions and preparation techniques infuence it as well. And a mix is almost never described with all of the little details.
* Unexpected inaccuracies: class of fy ash is not reported and superplasticizers (from the data) are from different manufacturers and of different chemical compositions and no other details.
* Limitations of the model

###Conclusions
*   With Linear Regression and Lasso we could produced a relatively accurate model with our dataset.
*   Linear Regression performs slightly better than Lasso for our dataset.
*   Through cross validation and iterative process we could determine which parameters we should use.
*   Excluding age from the feature by grouping datasets by age produced a much better performing model.

###Future Work
*   Model selection to select only the relevant predictor variables that can statistically explain concrete strength.
*   Check multicollinearity using VIF values.
*   Consider polynomial regression or neural networks to improve model prediction as done in Yeh’s article.
